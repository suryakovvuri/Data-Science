```python
import pandas as pd
import numpy as np
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_curve
```

```python
import math
%pylab inline
```

Reading the data set

```python
df = pd.read_csv("breast_cancer.csv")
```

```python
df
```

```python
df.describe()
```

Dropping the unwanted

```python
df = df.drop(['Unnamed: 0', 'id number'], axis=1)
```

malignant is dependent variable

```python
y = df.pop("malignant")
```

Training Set

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=.2, random_state=42)
```

```python
n_estimators = [300,400,500]
max_features = ['auto', 'sqrt','log2']
min_samples_split = [3,5,7]
```

```python
rfc = RandomForestClassifier(n_jobs=1)
```

```python
estimator = GridSearchCV(rfc,
                         dict(n_estimators=n_estimators,
                              max_features=max_features,
                              min_samples_split=min_samples_split
                              ), cv=None, n_jobs=-1)
```

```python
estimator.fit(X_train, y_train)
```

```python
estimator.best_estimator_
```

```python
best_rfc = estimator.best_estimator_
```

Finding Accuracy

```python
accuracy = accuracy_score(y_test, best_rfc.predict(X_test))
print "Accuracy is ", accuracy
```

```python
y_hat = best_rfc.predict(X_test)
```

```python
y_hat
```

```python
y_test
```

```python
correct = 0
total = y_test.shape[0]
for pred_val, truth_val in zip(y_hat, y_test):
    if pred_val == truth_val:
        correct +=1

print correct / float(total)
```

Precision and Recall

```python
print classification_report(y_test, best_rfc.predict(X_test))
```

precision model will correctly identify 96% maligiant cases, there is a 4% chance that non maligiant cases will be identified

recall means the people who are identified as maligiant, 96% are actualliy maligiant

```python
print confusion_matrix(y_test, best_rfc.predict(X_test))
```

K-fold Cross validation

```python
from sklearn import cross_validation
scores = cross_validation.cross_val_score(best_rfc, df, y, cv=10)
```

```python
scores
```

```python
mean_score = scores.mean()
std_dev = scores.std()
std_error = scores.std() / math.sqrt(scores.shape[0])
ci =  2.262 * std_error
lower_bound = mean_score - ci
upper_bound = mean_score + ci

print "Score is %f +/-  %f" % (mean_score, ci)
print '95 percent probability that if this experiment were repeated over and over the average score would be between %f and %f' % (lower_bound, upper_bound)
```
